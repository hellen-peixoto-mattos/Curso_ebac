## 1\. Preços `outliers` de diamante

```
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
```

```
# gráfico do exercício 1

diamonds = data[['price']]

with sns.axes_style('ticks'):
  grafico = sns.histplot(data=diamonds, x='price')
  grafico.set(title='Preços outliers de diamante', xlabel='Preço', ylabel='Frequência')
```

**Insight do gráfico 1**:

Analisando o gráfico de preços de diamantes, observo que a maior frequência de valores está em torno de **100** reais, enquanto a menor é de **17.550** reais. Essa disparidade sugere uma ampla gama de preços, com uma clara tendência para diamantes mais acessíveis. Essa observação pode ser atribuída a várias estratégias de vendas, como promoções ou diferenciações de mercado. Acredito que para uma interpretação mais aprofundada, é importante verificar outros fatores que influenciam os preços dos diamantes, como a qualidade, tamanho, pureza e origem. Dependendo das características, podem comandar preços mais altos, explicando **outliers** como o valor de **17.550** reais.

## 2\. Correlação entre o do preço do diamante com seu peso (`carat`) agrupados por sua transparêcia (`clarity`)

```
import seaborn as sns

data = sns.load_dataset("diamonds")

data. head()
```

```
diamonds = data[['price','carat','clarity']]

with sns.axes_style('whitegrid'):
  grafico = sns.scatterplot(data=diamonds, x='price', y='carat', hue='clarity', palette='bright')
  grafico.set(title='Correlação entre o preço do diamante com seu peso', xlabel='Preço', ylabel='Peso')
  grafico.get_legend().set_title('Transparência')
```

**Insight do gráfico 2**:

Percebi-se nos dados que independentemente da **transparência** e do peso dos diamantes, muitos deles têm um preço de venda comum de **17.500**. No entanto, ao aprofundar a análise, torna-se evidente que a clareza do diamante desempenha um papel significativo na determinação do preço. Notavelmente, diamantes classificados com a clareza **L1** parecem ser mais baratos, mesmo sendo mais pesados. Isso sugere que, embora tenham um tamanho maior, seu valor é atenuado por imperfeições e detalhes que afetam sua qualidade e seu preço no mercado. Por outro lado, diamantes com clareza **IF**, mesmo com pesos menores, tendem a ser mais caros. Isto pode ser atribuído à sua excepcional **transparência** e à escassez relativa desses diamantes no mercado.

## 3\. Distribuição contínua aproximada do peso (`weight`) de carros

```
import seaborn as sns

data = sns.load_dataset("mpg")
data.head()
```

```
# gráfico do exercício 3
titanic = data[['weight']]

with sns.axes_style('ticks'):
  grafico = sns.kdeplot(data=titanic, x='weight', fill=True)
  grafico.set(title='Gráfico de Densidade do Peso dos Carros', xlabel='Peso', ylabel='Densidade')
```

**Insight do gráfico 3**:

Visualizando o gráfico, percebi uma tendência para **carros** mais leves, sugerindo possível preferência por eficiência e desempenho. É crucial considerar as diversas influências que podem moldar os padrões de peso, como regulamentações de segurança, demanda do mercado e avanços tecnológicos. A faixa de peso comum é em torno de **2.300** quilos, mas há uma escassez relativa em torno de **5.800** quilos, e nenhum carro na faixa de **6.000** quilos. Isso pode indicar lacunas de mercado ou limitações nos dados disponíveis.
