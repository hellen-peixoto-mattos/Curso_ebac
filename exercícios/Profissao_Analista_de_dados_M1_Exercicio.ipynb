## 2\. Números

Preencha as células de código para preencher os valores de (A), (B) e (C) na tabela de ticket médio abaixo:

<br>

| Dia   | Valor Total Vendas | Qtd Total Vendas | Ticket Medio |
|-------|--------------------|------------------|--------------|
| 19/01 | (A)                | 3                | 320.52       |
| 20/01 | 834.47             | (B)              | 119.21       |
| 23/01 | 15378.12           | 5                | (C)          |

```
# (A)
qtd_vendas_19 = 3
tkt_19 = 320.52

vtv_19 = qtd_vendas_19 * tkt_19
print(vtv_19)
```

```
# (B)
vtv_20 = 834.47
tkt_20 = 119.21

qtd_vendas_20 = vtv_20 / tkt_20
print(qtd_vendas_20)
```

```
# (C)
vtv_23 = 15378.12
qtd_vendas_23 = 5

tkt_23 = vtv_23 / qtd_vendas_23
print(tkt_23)
```

## 3\. Strings

Aplique três métodos distintos na *string* abaixo, você pode conferir alguns métodos neste [link](https://www.w3schools.com/python/python_ref_string.asp):

```
cancao = 'Roda mundo, roda gigante, roda moinho, roda pião.'

string_1 = 'roda mundo'
string_2 = 'roda gigante'
string_3 = 'roda moinho'
string_4 = 'roda pião'

resultado_1 = string_1.upper()
resultado_2 = string_2.upper()
resultado_3 = string_3.upper()
resultado_4 = string_4.upper()

print(resultado_1, resultado_2, resultado_3, resultado_4)
```

```
string_1_2 = 'roda mundo'
string_2_2 = 'roda gigante'
string_3_2 = 'roda moinho'
string_4_2 = 'roda pião'

resultado_1 = string_1_2.replace('mundo', 'mundão')
resultado_2 = string_2_2.replace('gigante', 'pequena')
resultado_3 = string_3_2.replace('moinho', 'catavento')
resultado_4 = string_4_2.replace('roda', 'vira')

print(resultado_1, resultado_2, resultado_3, resultado_4)
```

```
string_1_3 = 'roda mundo'
string_2_3 = 'roda gigante'
string_3_3 = 'roda moinho'
string_4_3 = 'roda pião'

resultado_1 = string_1_3.count('o')
resultado_2 = string_2_3.count('o')
resultado_3 = string_3_3.count('o')
resultado_4 = string_4_3.count('o')
print(resultado_1, resultado_2, resultado_3, resultado_4)
```

```
noticia = 'Selic vai a 2,75% e supera expectativas; é a primeira alta em 6 anos.'

selic = noticia.find('2,75%')
selic_fim = selic + len('2,75%')
valor_selic = noticia[selic:selic_fim]

anos = noticia.find('6 anos')
valor_anos = noticia[anos:]

print(valor_selic)
print(valor_anos)
```
